## Model Deployment

After the model development and testing, we are now ready to deploy the model, and start serving.

### Model Registry

First step towards model deployment is to register the model. 
In EPIC, navigate to the Model Registry. In the Model Management screen, click on <font color='green'>Register New Model</font>  and register the model as shown below. Select <font color='green'>Browse</font> to choose the model directory path for the pickle file and the code directory for the scoring script. 

![Model Registry](Pictures/ModelRegistry.png)

### Model Deployment

Deploy the newly registered model by clicking the Deploy icon (jewel), under Actions, for the Income_Prediction model. This method automatically pre-selects the Income_Prediction model and its version from model registry to deploy.

![ModelDeployment](Pictures/ModelDeployment.png)

Ensure that the Deployment cluster status appears as Ready.

**NOTE:** Alternately, the other method to achieve the same result is to navigate to the Deployments screen and click on <font color='green'>Create Deployment</font> and enter a name for Deployment cluster and choose the model name and version from the dropdown. You may leave the rest of the configuration settings as default.

### Model Serving

To generate income prediction requests, we will use the **curl** command to post a prediction request. You could also use **Postman** tool, if already installed on your laptop. The below curl command should be copied to your laptop terminal window after making the two changes in the steps below. You can copy the curl command in your favorite text editor for editing or create a bash script to test.

Navigate to the Deployments screen and click on the <font color='green'>Income_Prediction</font> link. You will see a list all the nodes (containers) running for this deployment cluster with various roles. 

![ModelServing](Pictures/ModelServing.png)
Copy the Model Serving Load Balancer URL and paste it in the **curl** command shown below for the post request for model serving.

![AuthToken](Pictures/AuthToken.png)
Hover over and Click on the <font color='green'>Auth Token</font>, it gets copied to the clipboard. Paste it in the **curl** command in the header Auth-Token shown below.

curl --location -v -k --request POST '***your model serving URL here***/Income_Prediction/1/predict' \
--header 'X-Auth-Token: ***paste your copied Auth-Token here***' \
--header 'Content-Type: application/json' \
--data-raw '{
    "use_scoring": true,
    "scoring_args": {
        "age": 35,
        "workclass": "Self-emp-inc",
        "fnlwgt": 182148,
        "education": "Bachelors",
        "education_num": 13,
        "marital_status": "Married-civ-spouse",
        "occupation": "Exec-managerial",
        "relationship": "Husband",
        "race": "White",
        "sex": "Male",
        "capital_gain": 0,
        "capital_loss": 0,
        "hours_per_week": 60,
        "native_country": "United-States"
    }
}'

The output of this post should be as follows, which shows all the parameters passed via the API and prediction of salary being greater than 50K with about 81% confidence.
