## Lab 4: Deploy your App to Netlify

This Jupyter notebook will drive you through the following steps :

- Signup tp netlify
- Link your github repository to netlify
- Install netlify to your GitHub account
- Edit deployment setting
- Deploy your app

### Netlify?
Netlify is simply used for hosting static sites. more like GitHub pages. 

- They give you a custom URL for your project under netlify.com
- You can CNAME this to any domain you also own
- you can drag and drop a folder with all your HTML, CSS, JS, etc
- you can even connect a project to a GitHub project and have it automatically deploy with commits with even CI or just raw upload

### Signup

Signup to [Netlify](https://app.netlify.com/signup) using your GitHub account

![netlifysignup.png](Pictures/netlifysignup.png)

After you signup, you will be redirected to neftlify account site section.

### Link your github repository to netlify

Click on New site from git to link your github repository to netlify or simply drag and drop your site folder

Let's proceed with New Site from Git option

![netlifyGit.png](Pictures/netlifyGit.png)

Choose the Git provider where your siteâ€™s source code is hosted and select the repository to grant access to netlify.

![netlifyselectrepo.png](Pictures/netlifyselectrepo.png)

### Install netlify to your GitHub account
Once you select the repo, click on install to install netlify to your repository.

On next screen, select your repository to proceed.

![netlifyselectrepo1.png](Pictures/netlifyselectrepo1.png)

### Check Build settings and Deploy

On next screen, select check your build settings and click on Deploy.

![netlifydeploy.png](Pictures/netlifydeploy.png)

You can see your app deployement is in progress with random domain name. 

You can edit the default domain name by clicking on Domain Settings

![netlifydomainname.png](Pictures/netlifydomainname.png)

provide either existing domain name if you have one or you can enter a custom domain name EX:tss-grommet-studentXX.netlify.com

![netlifydomainname1.png](Pictures/netlifydomainname1.png)

![netlifydomainname2.png](Pictures/netlifydomainname2.png)

You can see your site url and click on it to go to your site.

![netlifydeployedsite.png](Pictures/netlifydeployedsite.png)

By default, auto publish will be turned on, if you make any changes to your github source code, netlify will auto deploy your site.

![netlifysite.png](Pictures/netlifysite.png)
