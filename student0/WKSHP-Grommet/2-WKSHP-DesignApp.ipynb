## How to add a component and edit its properties

Before proceeding further, let's see how to add a component and edit its properties on the Grommet designer site.

In the left sidebar, click on the + icon add a new component.

After adding a component, select that component on the left sidebar and edit its properties on the right sidebar.

**NOTE**

- To ignore the property value, simply make its value as "undefined".

- You can find documentation, reset, duplicate and delete options for each component on the right side top corner as shown below

![ResetDelete.PNG](Pictures/resetdelete.png)

## Step 0 - Import Existing Designs

### Change Theme

As you finished designing your app, let's see how to apply/change to a different theme.

By default, the grommet theme will be selected in the designer.grommet.io, and you will see all your buttons, links etc in violet color. To select a different theme, click on your design name "StudentXX", click on Configure, and then click on the URL under Theme section and change it to your desired theme EX: hpe and close the popup.

Now, you can see buttons colors are changed from violet (grommet theme) to green as per your theme selection (HPE).

![theme.png](Pictures/theme.png)


### Theme Mode

Grommet also offers you a couple of theme modes, light and dark mode.

if you don't want to view your app with too much brightness, select the dark mode and vice versa.

Click on your design name "StudentXX", Configure, and select the theme mode as either light or dark to see the difference.
